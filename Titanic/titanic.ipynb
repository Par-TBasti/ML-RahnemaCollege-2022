```python
import pandas
import numpy
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
traindataset = pandas.read_csv('train.csv')
```

```python
testdataset = pandas.read_csv('test.csv')
```

```python
traindataset.keys()
```

```python
testdataset.keys()
```

```python
traindataset.isna().sum()
```

```python
traindataset.Age.fillna(traindataset.Age.mean(), inplace=True )
```

```python
traindataset.Embarked.fillna('S', inplace=True)
```

```python
traindataset.isna().sum()
```

```python
traindataset = pandas.get_dummies(traindataset, columns=['Pclass', 'Sex'])
```

```python
train, val = train_test_split(traindataset)
```

```python
val.keys()
```

```python
train_features = ['Sex_female', 'Sex_male', 'Pclass_1', 'Pclass_2', 'Pclass_3','Survived']
```

```python
train = train[train_features]
```

```python
val = val[train_features]
```

```python
xtrain = train.drop('Survived', axis=1)
xval = val.drop('Survived', axis=1)
ytrain = train.Survived
yval = val.Survived
```

```python
lr = LogisticRegression()
```

```python
lr.fit(xtrain, ytrain)
```

```python
lr.predict(xval)
```

```python
lr.score(xval, yval)
```

```python
lr.score(xtrain, ytrain)
```
