```python
data = pandas.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
data["target"] = dataset.target
```

```python
data.describe()
```

```python
profile = ProfileReport(data)
```

```python
profile.to_file('Boston.html')
```

```python
data = data.sample(frac = 1)
```

```python
train, test = train_test_split(data, test_size=.2)
```

```python
train.shape, test.shape
```

```python
xtrain = train.LSTAT.values.reshape(-1, 1)
ytrain = train.target.values
xtest = test.LSTAT.values.reshape(-1, 1)
ytest = test.target.values
```

```python
lr = LinearRegression()
```

```python
lr.fit(xtrain, ytrain)
```

```python
lr.coef_, lr.intercept_
```

```python
yhatTrain = lr.predict(xtrain)
yhatTest = lr.predict(xtest)
yhat = lr.predict(data.LSTAT.values.reshape(-1, 1))
```

```python
fig, ax = matplotlib.pyplot.subplots(figsize = (6,6))
ax.set_xlabel('LSTAT')
ax.set_ylabel('Target')
ax.scatter(xtrain, ytrain)
ax.scatter(xtrain, yhatTrain)
```

```python
np.sqrt(mean_squared_error(ytrain, yhatTrain))
```

```python
np.sqrt(mean_squared_error(ytest, yhatTest))
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
mean_absolute_error(ytrain, yhatTrain), mean_absolute_error(ytest, yhatTest)
```

```python
r2_score(ytrain, yhatTrain), r2_score(ytest, yhatTest)
```
